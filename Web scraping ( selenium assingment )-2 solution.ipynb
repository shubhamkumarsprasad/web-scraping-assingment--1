# Q4 Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:

```python
search4=driver.find_element(By.XPATH,'/html/body/div/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
search4.click()
```

```python
brand=[]
product_design=[]
glass_price=[]
```

```python
start=0
end=3
for i in range(start,end):
    brand_tag=driver.find_elements(By.XPATH, '//div[@class="_2WkVRV"]')
    for i in brand_tag:
        brand.append(i.text)
    
    product_design_tag=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in product_design_tag:
        product_design.append(i.text)
        
    price_tag=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in price_tag:
        glass_price.append(i.text)
    next_button=driver.find_element(By.XPATH,'//a[@class="ge-49M _2Kfbh8"]')
    next_button.click()
    time.sleep(3)
```

```python
product_designs=product_design[0:100]
product_designs
```

# Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on TopQuotes
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
quote=driver.get('https://www.azquotes.com/')
```

```python
quote
```

```python
quotes=[]
author=[]
type_of_quote=[]
```

```python
start=0
end=10
for page in range(start,end):
    quote_tag=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quote_tag:
        quotes.append(i.text)
    author_tag=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in author_tag:
        auth=i.text
        author.append(auth)
    type_quote_tag=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in type_quote_tag:
        qo=i.text
        type_of_quote.append(qo)
    next_button=driver.find_element(By.XPATH,'//a[@class="active"]')
    next_button.click()
    time.sleep(3)
```

```python
authors=author[0:999]
authors
```

```python
quotes1=quotes[0:999]
quotes1
```

```python
types_of_quotes=type_of_quote[0:999]
types_of_quotes
```

```python
df8=pd.DataFrame({'quotes':quotes1,'author_name':authors,'type_of_quote':types_of_quotes})
df8
```
